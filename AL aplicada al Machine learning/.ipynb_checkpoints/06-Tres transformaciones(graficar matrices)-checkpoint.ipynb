```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
#graficando un circulo unitario
x = np.linspace(-1,1, 100000)#crea un vector entre -1 y 1 con 100000 puntos equidistantes
y = np.sqrt(1-(x**2))#funcion de un circulo x**2+y**y=r**2 despejando y y haciendo r=1
```

```python
print(x)
```

```python
print(y)
```

```python
#graficando
plt.plot(x,y,'green',alpha = 0.7)
plt.plot(x,-y,'green',alpha = 0.7)
```

```python
A = np.array([[3,7],[5,2]])
B = np.eye(2)
```

```python
Circulo_Transformado = A.dot(np.array([x,y]))
```

```python
plt.plot(Circulo_Transformado[0,:],Circulo_Transformado[1,:],'green',alpha = 0.7)
plt.plot(-Circulo_Transformado[0,:],-Circulo_Transformado[1,:],'green',alpha = 0.7)
#este pequeño ejercicio es paradeostrar como se transforma es espacio vectorial del circulo unitario
#y como se puede graficar este me diante el uso del producto punto y la funcion plt.plot
```
