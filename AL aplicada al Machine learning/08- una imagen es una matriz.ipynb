```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image #libreria para leer imagenes
from skimage import io # otra libreria para cargar imagenes y trabajar con imagenes
```

```python
plt.style.use('classic') 

imagen = Image.open("..\\imagenes\\imagen_ejemplo_frida_bredesen.jpg")



image=io.imread("..\\imagenes\\imagen_ejemplo_frida_bredesen.jpg")/255.0 # imread lee las imagenes con los pixeles codificados como enteros 
# en el rango 0-255. Por eso la convertimos a flotante y en el rango 0-1

print("- Dimensiones de la imagen:")
print(image.shape)
plt.imshow(image,vmin=0,vmax=1)
#plt.imshow(imagen)
print(imagen.getbands)
```

```python
### pasando a escala de grises
imagen_gr = imagen.convert('LA')#esta funcion devuelve la imagen en escala de grises con 2 canales L para la imagen en blanco y negro
                                #A para la transparencia de la imagen
                                #para imagenes en JPG la transparencia siempre es valores 255
plt.imshow(imagen_gr)
print(imagen_gr)
print(imagen_gr.getbands)
```

```python
numpy.set_printoptions(threshold = False)
```

```python
# import sys
# #import numpy
    #en la siguiente funcion se toma la banda 0 que es la que tiene la informacion L = escala de grices, recordar que el segundo canal es A= transparencia 
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)# esta funcion entrega los datosde forma flatten, osea organiza todo en un vector estirado
#shimg= np.array(imagen_gr)
# with np.printoptions(threshold=10000000):
#     print(imagen_mat)
print(imagen_mat)
print(imagen_mat.shape)
#print(shimg[1])
#plt.imshow(shimg[:,:,0])
```

```python
imagen_mat.shape =(imagen_gr.size[1], imagen_gr.size[0])#le da la forma de imagen_gr a imagen_mat 
                                                        #recordar que las imagenes se nombran columnasxfilas y las matrices filasxcolumnas
                                                        #por eso pone primero el dato size[1] que size[0]
print(imagen_mat)
```

```python
print(imagen_mat.shape)#verificamos el shape(forma) de la imagen pero leida como una matriz (FilasxColumnas)
```

```python
plt.imshow(imagen_mat,cmap = 'gray')
plt.show
```

```python
imagen_mat_2 = imagen_mat/10 #al dividir toda la imagen en 10 no se ve alterada la imagen, siempre va a comprar la relacion entre pixeles
print(imagen_mat_2)
plt.imshow(imagen_mat_2)
plt.show()
```

```python
plt.imshow(imagen_mat_2,cmap="gray")#aqui como indicamos que es una escala de grices se ve normal
plt.show
```

```python
# en machine learning se tiende a normalizar la informacion para que sea mas facil realizar los calculos
print(np.max(imagen_mat_2))
print(np.max(imagen_mat))
```
