```python
%matplotlib notebook #este permite mover los graficos
import numpy as np
import matplotlib.pyplot as plt

```

```python
#definiendo vectores
v1 = np.array([1,1])
v2 = np.array([-1, -1])# para este caso vamos a ver q aunque estos vecores esten en R2, su subespacio esta solo en R1 
```

```python
#este codigo sirve para dibujar un rango en 2d de los vectores v1 y v2
for a in range (-10,10):
    for b in range(-10,10):
            plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a +v2[1]*b,
                       marker = '.',
                       color = "orange")
            
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.axvline(x=0,color='grey')
plt.axhline(y=0,color='grey')
plt.show()
```

```python
v1 = np.array([1,0])
v2 = np.array([2,-3])#con estos 2 vectores si podemos formar un subespacio R2
```

```python
#reutilizamos el codigo anterior
for a in range (-10,10):
    for b in range(-10,10):
            plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a +v2[1]*b,
                       marker = '.',
                       color = "orange")
            
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.axvline(x=0,color='grey')
plt.axhline(y=0,color='grey')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
v1 = np.array([1,0,0])
v2 = np.array([2,-3,0])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')#agrega los ejes que vamos a utilizar

for a in range (-10,10):
    for b in range(-10,10):
            ax.scatter(v1[0]*a + v2[0]*b, v1[1]*a +v2[1]*b,#hacemos modificaciones apra trabajarlo en 3D
                       v1[2]*a + v2[2]*b, #hacemos modificaciones apra trabajarlo en 3D
                       marker = '.',
                       color = "orange")
            
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')


plt.show()
```
